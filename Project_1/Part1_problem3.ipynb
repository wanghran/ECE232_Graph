### 3. Create a modified preferential attachment model that penalizes the age of a node
### (a)
The degree distribution of the network is as below.

```r
g3a = aging.prefatt.game (1000, 1, -1, m = 1, aging.bin = 300,
     out.dist = NULL, out.seq = NULL, out.pref = TRUE,
     directed = FALSE, zero.deg.appeal = 1, zero.age.appeal = 0,
     deg.coef = 1, age.coef = 1, time.window = NULL)
degree_graph = degree(g3a)
degree_dist = hist(x = degree_graph, breaks = seq(from = min(degree_graph), to = max(degree_graph), by=1), main = "Degree Distribution for Evolving Graph", xlab = "Number of Degrees")
```

Fit the distribution to power law and get the result exponent is 
4.68233571652074. Details are as below.

```r
fit_power_law(degree_graph)
```
