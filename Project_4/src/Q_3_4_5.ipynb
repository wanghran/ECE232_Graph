```r
library('igraph')
```

```r
g <- read_graph('../output/edge_list.txt', format='ncol')
```

```r
result = sort(page_rank(g, directed = TRUE)$vector, decreasing = TRUE)
```

```r
print(result)
```

```r
top_10 = attr(result[1:10], 'names')
```

```r
# print(top_10)
```

```r
degrees = degree(g, v = V(g), mode = c("in"))
# print(degrees)
```

```r
in_degrees = seq(0)
for (i in seq(1,length(degrees))){
    name = attr(degrees[i], "names")
    for (actor in top_10){
        if (name == actor){
            print(degrees[i])
            in_degrees = c(in_degrees, degrees[i])
        }
    }
}
```
